## filling missing value

```python
from sklearn.impute import SimpleImputer
```

```python
imputer = SimpleImputer(strategy='median')
```

```python
imputer.fit(df)
```

```python
imputer.statistics_
```

```python
x = imputer.transform(df)
```

```python
copy_data = pd.DataFrame(x, columns=df.columns)
copy_data
```

```python
copy_data.shape
```

```python
copy_data.count()
```

```python
from sklearn.model_selection import StratifiedShuffleSplit
split = StratifiedShuffleSplit(n_splits=1, test_size=0.2,random_state=42)
for train_index, test_index in split.split(copy_data, copy_data['CHAS']): 
    train_set = copy_data.loc[train_index]
    test_set = copy_data.loc[test_index]
```

```python
train_set
```

```python
data = train_set.drop(['MEDV'],axis=1)
dataset_labels = train_set['MEDV'].copy()
```

```python
test_set
```

# :: Trying to Build Pipeline::

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
my_pipe = Pipeline([
    ('imputer', SimpleImputer(strategy='median')),
    ('stdclr', StandardScaler()),
    
])
```

```python
dataset = my_pipe.fit_transform(data)
```

```python
dataset
```

## model selection

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(dataset,dataset_labels)
```
